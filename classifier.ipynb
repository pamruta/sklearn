```python
# load sample dataset

from sklearn.datasets import load_wine
wine = load_wine()
x, y = wine.data, wine.target
```

```python
# split dataset into train and test

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3)
```

```python
# run predictions using decision tree

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier().fit(train_x, train_y)
model.score(test_x, test_y)
```

```python
# run predictions using SVM

from sklearn.svm import SVC
model = SVC().fit(train_x, train_y)
model.score(test_x, test_y)
```

```python
# run predictions using KNN

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3).fit(train_x, train_y)
model.score(test_x, test_y)
```
